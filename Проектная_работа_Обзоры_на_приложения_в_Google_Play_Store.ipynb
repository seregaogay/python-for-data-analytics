# 1.Подготовьте список самых популярных приложений по категориям

```python
 df['score'] = df['Maximum Installs'] * df['Rating'] / 1000000
```

```python
df_top = pd.DataFrame()

for i in df.Category.unique():
    df_top = df_top.append(df[df['Category'] == i].sort_values('score', ascending=False).head(3))
```

```python
df_top[['App Name', 'Category', 'score']].sort_values(['Category', 'score'], ascending=False)
```

# 2.Какие приложения самые дорогие (и категории)

```python
df['income'] = df['Price'] * df['Maximum Installs']/1000000
```

```python
df[['App Name', 'Category', 'income']].sort_values('income', ascending=False)
```

```python
df[['App Name', 'Category', 'income']].groupby('Category').agg('sum')\
.sort_values('income', ascending=False)
```

# 3.Какие категории содержат самое большое количество приложений без отзывов
